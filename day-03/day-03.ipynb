```python
file1 = open('day03-input.txt', 'r')
lines = file1.readlines()

# PART 1
item_types = []
count = 0

for i in lines:
    mid_index = int((len(i)-1) / 2)
    first_compartment = i[0:mid_index]
    second_compartment = i[mid_index:]

    for l in range(len(first_compartment)):
        if first_compartment[l] in second_compartment:
            item_type = first_compartment[l]
            item_types.append(item_type)
            break
```

```python
item_types = sorted(item_types, key=str.swapcase)
item_types
```

```python
# (p), 38 (L), 42 (P), 22 (v), 20 (t), and 19 (s)
# item_types = ['p', 'L','P', 'v', 't', 's']

priority_array = []

for x in item_types:
    if x.islower():
        p = ord(x) - 96
    elif x.isupper():
        p = ord(x) - 64 + 26

    priority_array.append(p)

sum(priority_array) #<- too high so gonna remove duplicates
```

```python
file1 = open('day03-input.txt', 'r')
lines = file1.readlines()

# PART 2
group_badges = []

for i in range(0, len(lines), 3):
    line1 = lines[i]
    line2 = lines[i+1]
    line3 = lines[i+2]

    for l in range(len(line1)):
        if (line1[l] in line2) and (line1[l] in line3):
            print(line1[l])
            group_badges.append(line1[l])
            break
```

```python
badge_priority_array = []

for x in group_badges:
    if x.islower():
        p = ord(x) - 96
    elif x.isupper():
        p = ord(x) - 64 + 26

    badge_priority_array.append(p)

sum(badge_priority_array) 
```
