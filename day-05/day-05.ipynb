```python
file1 = open('day05-input.txt', 'r')
lines = file1.readlines()

# PART 1

#             [C]         [N] [R]    
# [J] [T]     [H]         [P] [L]    
# [F] [S] [T] [B]         [M] [D]    
# [C] [L] [J] [Z] [S]     [L] [B]    
# [N] [Q] [G] [J] [J]     [F] [F] [R]
# [D] [V] [B] [L] [B] [Q] [D] [M] [T]
# [B] [Z] [Z] [T] [V] [S] [V] [S] [D]
# [W] [P] [P] [D] [G] [P] [B] [P] [V]
#  1   2   3   4   5   6   7   8   9 

arr1 = ['W', 'B', 'D,', 'N', 'C', 'F', 'J']
arr2 = ['P', 'Z', 'V,', 'Q', 'L', 'S', 'T']
arr3 = ['P', 'Z', 'B,', 'G', 'J', 'T']

arr4 = ['D', 'T', 'L,', 'J', 'Z', 'B', 'H', 'C']
arr5 = ['G', 'V', 'B,', 'J', 'S']
arr6 = ['P', 'S', 'Q,']

arr7 = ['B', 'V', 'D,', 'F', 'L', 'M', 'P', 'N']
arr8 = ['P', 'S', 'M,', 'F', 'B', 'D', 'L', 'R']
arr9 = ['V', 'D', 'T,', 'R']

stacks = [arr1, arr2, arr3, arr4, arr5, arr6, arr7, arr8, arr9]

count = 0

def moveCrate(reps, from_stack, to_stack):
    for i in range(reps):
        stacks[to_stack].append(stacks[from_stack].pop())
    printStacks()

def moveMultipleCrate(reps, from_stack, to_stack):
    helper_arr = []

    for i in range(reps):
        helper_arr.append(stacks[from_stack].pop())
    
    print("HELPER", helper_arr)
    helper_arr.reverse()
    for j in helper_arr:
        stacks[to_stack].append(j)
    printStacks()

def printStacks():
    for stack in stacks:
        print(stack)
    print("---------------------------")

print("STARTING")
printStacks()
```
